```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
var=sns.load_dataset("tips")
var
```

```python
sns.violinplot(x="day",y="total_bill",data=var)
```

```python
sns.violinplot(x="day",y="total_bill",data=var,hue="time")
```

```python
sns.violinplot(x="day",y="total_bill",data=var,linewidth=2)
```

```python
sns.violinplot(x="day",y="total_bill",data=var,hue="time",palatte="Accent")
```

```python
sns.violinplot(x="time",y="total_bill",data=var,order=["Dinner","Lunch"])
```

```python
sns.violinplot(x="time",y="total_bill",data=var,order=["Dinner","Lunch"],color='r')
```

```python
sns.violinplot(x="day",y="total_bill",data=var,split=True)
```
