```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset("penguins").head(30)
```

```python
df
```

```python

sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=df)
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=df)

plt.xlabel("Bill Length MM")
plt.ylabel("Bill Depth MM")
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=df,hue="sex")

plt.xlabel("Bill Length MM")
plt.ylabel("Bill Depth MM")
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=df,hue="sex",size="sex")

plt.xlabel("Bill Length MM")
plt.ylabel("Bill Depth MM")
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=df,hue="sex",size="sex",sizes=[20,30])

plt.xlabel("Bill Length MM")
plt.ylabel("Bill Depth MM")
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=df,hue="sex",size="sex",sizes=[20,30])
plt.xlabel("Bill Length MM")
plt.ylabel("Bill Depth MM")
plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=df,hue="sex",size="sex",sizes=[20,30],palette="binary_r",alpha=1)
plt.xlabel("Bill Length MM")
plt.ylabel("Bill Depth MM")
plt.show()
```

```python
marker_list={"Male":"x","Femae":"+"} 
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=df,hue="sex",size="sex",sizes=[20,30],
                 palette="binary_r",alpha=1,markers=marker_list)
plt.xlabel("Bill Length MM")
plt.ylabel("Bill Depth MM")
plt.show()
```
