```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset("penguins")
df
```

```python
sns.barplot(x=df.island,y=df.bill_length_mm)
plt.show()
```

```python
sns.barplot(x=df.island,y=df.bill_length_mm,hue=df.sex)
plt.show()
```

```python
order_1=["Dream","Torgersen","Biscoe"]
sns.barplot(x=df.island,y=df.bill_length_mm,hue=df.sex,order=order_1)
plt.show()
```

```python
order_1=["Dream","Torgersen","Biscoe"]
sns.barplot(x=df.island,y=df.bill_length_mm,hue=df.sex,
            order=order_1,hue_order=["Female","Male"])
plt.show()
```

```python
order_1=["Dream","Torgersen","Biscoe"]
sns.barplot(x=df.island,y=df.bill_length_mm,hue=df.sex,
            order=order_1,hue_order=["Female","Male"],ci=50)
plt.show()
```

```python
order_1=["Dream","Torgersen","Biscoe"]
sns.barplot(x=df.island,y=df.bill_length_mm,hue=df.sex,
            order=order_1,hue_order=["Female","Male"],palette="Accent")
plt.show()
```

```python
order_1=["Dream","Torgersen","Biscoe"]
sns.barplot(x=df.island,y=df.bill_length_mm,hue=df.sex,
            order=order_1,hue_order=["Female","Male"],palette="Accent",saturation=50)
plt.show()
```
