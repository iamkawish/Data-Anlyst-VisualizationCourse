```python
df1=sns.load_dataset('penguins')
df1
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=df1,hue='sex',size=10,style='sex',
             palette='Accent',markers=["o","*"],dashes=False)
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=df1,hue='sex',size=10,style='sex',
             palette='Accent',markers=["o","*"],dashes=True,legend='full')
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=df1,hue='sex',size=10,style='sex',
             palette='Accent',markers=["o","*"],dashes=True,legend='full')
plt.grid()
plt.title("Python Course")
plt.xlabel("Age of Student")
plt.ylabel("Height pf Student")

plt.show()
```
